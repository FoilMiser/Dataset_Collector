# Dataset Collector Quickstart


This notebook inspects pipeline configuration using repo data.


```python
from pathlib import Path

repo_root = Path.cwd()
if not (repo_root / "configs").exists():
    repo_root = repo_root.parent

import yaml

config_path = repo_root / "configs" / "pipelines.yaml"
config = yaml.safe_load(config_path.read_text())

pipelines = config["pipelines"]
print(f"Pipelines registered: {len(pipelines)}")
print("Sample domains:", ", ".join(sorted(pipelines.keys())[:5]))
```
