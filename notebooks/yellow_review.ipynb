# Yellow Review Exploration


Locate pipelines that require yellow screening.


```python
from pathlib import Path

repo_root = Path.cwd()
if not (repo_root / "configs").exists():
    repo_root = repo_root.parent

import yaml

config = yaml.safe_load((repo_root / "configs" / "pipelines.yaml").read_text())

pipelines = config["pipelines"]
with_yellow = [
    name
    for name, info in pipelines.items()
    if info.get("knobs", {}).get("yellow_screen_module") or info.get("yellow_screen")
]

print("Pipelines with yellow screen modules:")
for name in sorted(with_yellow):
    print("-", name)
```
