# Content Checks Validation


Validate the denylist configuration against its schema.


```python
from pathlib import Path

repo_root = Path.cwd()
if not (repo_root / "configs").exists():
    repo_root = repo_root.parent

import json
import yaml
from jsonschema import Draft7Validator

license_map = yaml.safe_load((repo_root / "configs" / "common" / "license_map.yaml").read_text())
schema = json.loads((repo_root / "schemas" / "license_map.schema.json").read_text())

validator = Draft7Validator(schema)
errors = sorted(validator.iter_errors(license_map), key=lambda e: e.path)

if errors:
    raise ValueError(f"Schema validation errors: {errors}")

print("License map validated successfully.")
```
