# Custom Pipeline Walkthrough


Inspect a pipeline entry and discover its targets file.


```python
from pathlib import Path

repo_root = Path.cwd()
if not (repo_root / "configs").exists():
    repo_root = repo_root.parent

import yaml

config = yaml.safe_load((repo_root / "configs" / "pipelines.yaml").read_text())

pipeline = config["pipelines"]["physics"]
print("Pipeline id:", pipeline["pipeline_id"])
print("Targets path:", pipeline["targets_path"])

targets_path = repo_root / pipeline["targets_path"]
print("Targets file exists:", targets_path.exists())
```
