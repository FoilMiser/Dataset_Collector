# Production Deployment Checklist


Inspect the production profile configuration.


```python
from pathlib import Path

repo_root = Path.cwd()
if not (repo_root / "configs").exists():
    repo_root = repo_root.parent

import yaml

profile = yaml.safe_load((repo_root / "configs" / "profiles" / "production.yaml").read_text())

print("Production profile keys:", ", ".join(sorted(profile.keys())))
```
